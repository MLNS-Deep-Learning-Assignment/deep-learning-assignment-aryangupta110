```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
train_data = np.load('data0.npy')
train_labels = np.load('lab0.npy')
train_data = train_data / 255.0
X_train, X_val, y_train, y_val = train_test_split(train_data, train_labels, test_size=0.2, random_state=42)
```

```python
model = tf.keras.models.load_model('preliminary_cnn_model.h5')

for i in range(10):
    plt.imshow(X_val[i].squeeze(), cmap='gray')  
    plt.show()
    print("True Label:", y_val[i])
    sample = np.expand_dims(X_val[i], axis=0)  
    prediction = model.predict(sample)
    print("Model Prediction:", prediction[0][0])
```
